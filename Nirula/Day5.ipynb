```python
from matplotlib import style
style.use('ggplot')
#plt.plot(x,y)
days=[1,2,3,4,5]
temp=[36,35,37.7,38,40.2]
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=10)
plt.plot(days,temp,color='g',marker='o',
         linewidth=3)
plt.show()
```

```python
vij_temp=[40.5,43.5,38.5,39.2,38]
plt.plot(days,vij_temp,'m*--',linewidth=5,label='vij')
plt.plot(days,temp,color='g',marker='o',linewidth=4,
        label='gnt')
plt.title('Gnt-Temp Vs Vij-Temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.legend(loc=4)
plt.grid(color='r',linestyle='-.',linewidth=2)
plt.show()
```

```python
#plotting an equation
import numpy as np
#y as random array
y=np.random.randint(10,30,10)
x=y**2+4
plt.plot(y,'m*--')
plt.plot(x,y,'yo-.')
plt.show()
```

```python
#histogram works similar to bar 
# data represented in bins form
std1=np.random.randint(18,45,100)
#std1
std2=np.random.randint(15,40,100)
#std2
plt.hist(std1,bins=50)
plt.xlabel('age category')
plt.ylabel("no.of students")
plt.show()
```

```python
# pie plot
x=[56,25,30,24,30] # making the regions
plt.pie(x)
plt.show()
```

```python
#adding labels
fb=['average','good','above average','Excellent','None']
plt.pie(x,labels=fb) # assigning labels of fb to x
plt.show()
```

```python
#adding colors to regions of pie
plt.pie(x,labels=fb,
        colors=['green','cyan','blue','pink','red'])
plt.title("Sort by Responses")
plt.show()
```

```python
ex=[0,1,0,1,0]
plt.pie(x,labels=fb,
        colors=['green','cyan','blue','pink','red'],
       autopct="%.f%%",shadow=True,startangle=180,
       explode=ex) # excluding part
plt.title("Sort by Responses")
plt.show()
```
